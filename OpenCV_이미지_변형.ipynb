```
import cv2
image = cv2.imread('cat.jpg') # image라는 변수에 해당 객체 저장

# 픽셀 수 및 이미지 크기 확인
print(image.shape) # 특정한 리스트 배열 객체의 shape과 크기 측정
print(image.size)

# 이미지 Numpy 객체의 특정 픽셀을 가리킵니다.
px = image[100, 100] # image에서 특정 x, y축에 해당하는 픽셀 가져오기

# B, G, R 순서로 출력됩니다.
# 단 Gray Scale인 경우에는 B, G, R로 구분되지 않습니다.
print(px)

# R 값만 출력하기
print(px[2])
```

```
import cv2
import matplotlib.pyplot as plt
import time

image = cv2.imread('cat.jpg')

start_time = time.time()
for i in range(0, 100): # 개별적인 픽셀에 일일이 접근해서 값 변경 가능
  for j in range(0, 100):
    image[i, j] = [255, 255, 255]
print("--- %s seconds ---" %(time.time() - start_time))

start_time = time.time()
image[0:100, 0:100] = [0, 0, 0] # 범위를 지정해서 값 변경 -> 이중for문을 이용하는 것보다 슬라이싱 연산을 이용하는 것이 시간 효율 good
print("---%s seconds ---" %(time.time() - start_time))

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('cat.jpg')

# Numpy Slicing: ROI 처리 가능
roi = image[100:150, 100:150]

image[50:100, 50:100] = roi

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('cat.jpg')
image[:, :, 2] = 0
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```
