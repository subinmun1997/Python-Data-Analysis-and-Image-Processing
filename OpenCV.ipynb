```
import cv2
import matplotlib.pyplot as plt

# OpenCV: [B,G,R] / Matplotlib: [R,G,B]
img_basic = cv2.imread('cat.jpg', cv2.IMREAD_COLOR) # 파일을 읽어서 numpy 객체로 생성
plt.imshow(cv2.cvtColor(img_basic, cv2.COLOR_BGR2RGB)) # matplotlib를 이용하여 파일 출력
plt.show()

img_basic = cv2.cvtColor(img_basic, cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(img_basic, cv2.COLOR_GRAY2RGB))
plt.show()
```
